```python
df=df.drop('Currency',axis=1)
```

```python
df=df.drop('City',axis=1)
```

```python
sns.scatterplot(x=df['Cuisines'],y=df['Aggregate rating'])
```

```python
df['Cuisines'].value_counts()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer()
cuisine_encoded = pd.DataFrame(mlb.fit_transform(df['Cuisines']), columns=mlb.classes_)
```

```python
df = pd.concat([df, cuisine_encoded],axis=1)
```

```python
df=df.drop(columns=['Cuisines'],axis=1)
```

```python
df['Is delivering now']=df['Is delivering now'].apply(lambda x: 1 if x=='No' else 0)
```

```python
df['Switch to order menu']=df['Switch to order menu'].apply(lambda x: 1 if x=='No' else 0)
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df=df.astype(int)
```

```python
dff=df.corr()['Aggregate rating'].abs()
```

```python
dff.sort_values(ascending=False).nlargest(30)
```

```python
point=0.1
top=dff[dff>=point].index
```

```python
df_top=df[top]
```

```python
df_top
```

```python
x=df_top.drop('Aggregate rating',axis=1)
y=df_top['Aggregate rating']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
print('r2_score',r2_score(y_test,y_pred))
print('mae',mean_absolute_error(y_test,y_pred))
```

```python
train_r2 = r2_score(y_train, lr.predict(x_train)) 
test_r2 = r2_score(y_test, lr.predict(x_test))

print(f'Training RÂ²: {train_r2:.3f}')
print(f'Test RÂ²: {test_r2:.3f}')
```

```python
rf=RandomForestRegressor(n_estimators=25,max_depth=7,random_state=15)
```

```python
rf.fit(x_train,y_train)
```

```python
r_pred=rf.predict(x_test)
```

```python
print('r2_score',r2_score(y_test,r_pred))
print('mae',mean_absolute_error(y_test,r_pred))
```

```python
#Hence RandomForestRegressor give more accuracy and less error as compare to other model.
```
